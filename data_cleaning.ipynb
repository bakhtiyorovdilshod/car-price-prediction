```python
def get_pozitsiya(name):
  name_list = name.split()
  if 'pozitsiya' in name_list:
    numbers = [val for val in name_list if val.isdigit()]
    if len(numbers) != 0:
      return numbers[-1]
    return 1
  return 1
```

```python
copied_df['pozitsiya'] = copied_df['title'].apply(get_pozitsiya)
```

```python
copied_df
```

```python
def get_name(title):
  title_list = title.split()
  if 'Chevrolet' in title_list:
    title_list.remove('Chevrolet')
  if 'pozitsiya' in title_list:
    numbers = [val for val in title_list if val.isdigit()]
    if len(numbers) != 0:
      title_list.remove(numbers[-1])
    title_list.remove('pozitsiya')
  if 'DOHC' in title_list:
    title_list.remove('DOHC')
  if 'EVRO' in title_list:
    title_list.remove('EVRO')
  if 'SOHC' in title_list:
    title_list.remove('SOHC')
  if 'Gaz-benzin' in title_list:
    title_list.remove('Gaz-benzin')
  if len(title_list) != 0:
    return ''.join(title_list)
  print(title_list)
  return ''
```

```python
copied_df['name'] = copied_df['title'].apply(get_name)
```

```python
copied_df['name'] = copied_df['name'].apply(lambda x: x.replace(',', ''))
```

```python
def is_float(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
x = "1.5 l, Benzin, Qora, Sedan, Uzatish qutisi Avtomat, Yengil qotishmali disklar, Lyuk, Yukxona, Ksenon, Tashqi chiroq pr…"
y = x.split()
if 'Uzatish' in y and 'qutisi' in y:
  ind = y.index('qutisi')
  print(y[ind+1])
```

```python
def get_transmission(description):
  description_list = description.split()
  if 'Uzatish' in description_list and 'qutisi' in description_list:
    return description_list[description_list.index('qutisi') + 1]
  return 'Mexanika'
```

```python
copied_df['engine_size'] = copied_df['description'].apply(lambda x: x.split()[0] if is_float(x.split()[0]) else 0)
```

```python
copied_df
```

```python
copied_df['transmission'] = copied_df['description'].apply(get_transmission)
```

```python
copied_df
```

```python
copied_df['transmission'] = copied_df['transmission'].apply(lambda x: x.replace(',', '') if isinstance(x, str)  else x)
```

```python
def get_fuel_type(text):
  text_list = [i.replace(',', '') for i in text.split() if i]
  if 'Benzin' in text_list:
    return 'Benzin'
  if 'Gaz-benzin' in text_list:
    return 'Gaz-benzin'
  if 'Gaz-Benzin' in text_list:
    return 'Gaz-benzin'
  if 'Elektr' in text_list:
    return 'Elektr'
```

```python
copied_df['fuel_type'] = copied_df['description'].apply(get_fuel_type)
```
